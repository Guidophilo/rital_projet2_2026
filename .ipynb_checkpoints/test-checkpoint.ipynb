```python
qrels_trec9 = pd.read_csv(                                                    
    "qrels.trec9.main_web.gz",                             
    sep=" ", header=None,                                                     
    names=["qid", "iteration", "docno", "relevance"]                          
)                                                                             
                                                                                
  # Topics déjà chargés, vérifier                                               
print(topics_trec9.head(3))
print(qrels_trec9.head(3))                                                    
print("Nb queries:", qrels_trec9["qid"].nunique())  
```

```python
# Pour chaque query, trier les documents par pertinence décroissante          
pseudo_results = []
                                                                                
for qid in qrels_trec9["qid"].unique():
    df_q = qrels_trec9[qrels_trec9["qid"] == qid].copy()                      
    df_q = df_q.sample(frac=1, random_state=42).reset_index(drop=True)        
    df_q["rank"] = df_q.index + 1
    df_q["score"] = (len(df_q) - df_q.index).astype(float)                    
    pseudo_results.append(df_q[["qid", "docno", "rank", "score"]])            
                                                                                
pseudo_df = pd.concat(pseudo_results, ignore_index=True)                      
print(pseudo_df.shape)
print(pseudo_df.head(5)) 
```

```python
pseudo_df["qid"] = pseudo_df["qid"].astype(str)
```

```python
from src.utils import results_to_doclist, results_to_score_dict,ranking_to_run_df                                                             
from src.subqueries import generate_dummy_subqueries                          
from src.importance import importance_uniform                                 
from src.xquad import xquad_rerank                                            
  
  # Test sur query 451                                                          
qid = "451"       
query = topics_trec9[topics_trec9["qid"] == qid]["query"].values[0]           
df_q = pseudo_df[pseudo_df["qid"] == qid].copy()
                                                                                
docs_ranked = results_to_doclist(df_q)
r_query = results_to_score_dict(df_q)                                         
                                                                                
  # Dummy subqueries
subqueries = generate_dummy_subqueries(query)                                 
                  
  # Simuler r_sub : mêmes scores que r_query pour chaque sous-requête           
r_sub = {doc: {sq: r_query.get(doc, 0) for sq in subqueries} for doc in docs_ranked}                                                                  
                  
  # Importance uniforme                                                         
imp = importance_uniform(subqueries)

  # Lancer xQuAD
result = xquad_rerank(docs_ranked, subqueries, r_query, r_sub, imp, tau=20,omega=0.5)                                                                    
  
print("Query:", query)                                                        
print("Baseline top5 :", docs_ranked[:5])
print("xQuAD top5    :", result[:5])
```

```python
all_results = []                                                              
                                                                                
for _, row in topics_trec9.iterrows():                                        
    qid = row["qid"]                                                          
    query = row["query"]

    df_q = pseudo_df[pseudo_df["qid"] == qid].copy()                          
    if df_q.empty:
        continue                                                              
                  
    docs_ranked = results_to_doclist(df_q)
    r_query = results_to_score_dict(df_q)
    subqueries = generate_dummy_subqueries(query)                             
    r_sub = {doc: {sq: r_query.get(doc, 0) for sq in subqueries} for doc in docs_ranked}                                                                  
    imp = importance_uniform(subqueries)
    result = xquad_rerank(docs_ranked, subqueries, r_query, r_sub, imp,tau=100, omega=0.5)

    all_results.append(ranking_to_run_df(qid, query, result))                 
  
run_xquad_trec9 = pd.concat(all_results, ignore_index=True)                                                       
print("xQuAD run:", run_xquad_trec9.shape)                                    
print(run_xquad_trec9.head(3)) 
```

```python
qrels_for_eval = qrels_trec9[["qid", "docno", "relevance"]].copy()            
qrels_for_eval.columns = ["qid", "docno", "label"]                            
qrels_for_eval["qid"] = qrels_for_eval["qid"].astype(str)                     
                                                                                
pseudo_df["qid"] = pseudo_df["qid"].astype(str)                               
run_xquad_trec9["qid"] = run_xquad_trec9["qid"].astype(str)                   
                                                                                
metrics = ["map", "ndcg_cut_10", "P_10"]                                      
  
res_baseline = pt.Evaluate(pseudo_df, qrels_for_eval, metrics=metrics)        
res_xquad = pt.Evaluate(run_xquad_trec9, qrels_for_eval, metrics=metrics)
                                                                                
print("Baseline :", res_baseline)
print("xQuAD    :", res_xquad)
```

### multi requête

```python
subset_topics = topics.head(5).copy()
print("Nombre de requêtes dans le sous-ensemble :", len(subset_topics))
subset_topics
```

```python
all_runs = {
    "baseline_bm25": [],
    "xquad_dummy_uniform": [],
    "xquad_dummy_n": [],
    "xquad_dummy_redde": [],
    "xquad_dummy_crcs": [],
    "xquad_kmeans_uniform": [],
    "xquad_kmeans_n": [],
    "xquad_kmeans_redde": [],
    "xquad_kmeans_crcs": [],
}

skipped_kmeans = []

for qid, query in subset_topics[["qid", "query"]].itertuples(index=False):
    print(f"Traitement qid={qid}")

    df_q = bm25_results[bm25_results["qid"] == qid].copy()
    docs_ranked = results_to_doclist(df_q)
    r_query = results_to_score_dict(df_q)

    # Baseline
    all_runs["baseline_bm25"].append(
        ranking_to_run_df(qid, query, docs_ranked[:100])
    )

    # -------------------------
    # Version dummy
    # -------------------------
    subqueries_dummy = generate_dummy_subqueries(query)

    _, r_sub_dummy, n_qi_dummy = prepare_subquery_scores(
        index_ref=index_ref,
        qid=qid,
        subqueries=subqueries_dummy,
        wmodel="BM25",
        num_results=1000,
    )

    importances_dummy = compute_all_importances(
        subqueries=subqueries_dummy,
        docs_ranked=docs_ranked,
        r_query=r_query,
        r_sub=r_sub_dummy,
        n_qi=n_qi_dummy,
        tau=100,
    )

    xquad_dummy = run_xquad_variants(
        docs_ranked=docs_ranked,
        subqueries=subqueries_dummy,
        r_query=r_query,
        r_sub=r_sub_dummy,
        importances=importances_dummy,
        tau=100,
        omega=0.5,
    )

    all_runs["xquad_dummy_uniform"].append(ranking_to_run_df(qid, query, xquad_dummy["uniform"]))
    all_runs["xquad_dummy_n"].append(ranking_to_run_df(qid, query, xquad_dummy["n"]))
    all_runs["xquad_dummy_redde"].append(ranking_to_run_df(qid, query, xquad_dummy["redde"]))
    all_runs["xquad_dummy_crcs"].append(ranking_to_run_df(qid, query, xquad_dummy["crcs"]))

    # -------------------------
    # Version KMeans
    # -------------------------
    df_q_top = df_q.head(100).copy()
    docs_with_text = df_q_top.merge(corpus_df, on="docno", how="left")

    subqueries_kmeans = build_subqueries_from_kmeans(
        docs_df=docs_with_text,
        text_col=text_col,
        n_clusters=5,
        top_k_terms=5,
        max_features=5000,
        random_state=42,
    )

    if len(subqueries_kmeans) == 0:
        skipped_kmeans.append(qid)
        print(f"  -> KMeans vide pour qid={qid}, requête ignorée pour la partie kmeans")
        continue

    _, r_sub_kmeans, n_qi_kmeans = prepare_subquery_scores(
        index_ref=index_ref,
        qid=qid,
        subqueries=subqueries_kmeans,
        wmodel="BM25",
        num_results=1000,
    )

    importances_kmeans = compute_all_importances(
        subqueries=subqueries_kmeans,
        docs_ranked=docs_ranked,
        r_query=r_query,
        r_sub=r_sub_kmeans,
        n_qi=n_qi_kmeans,
        tau=100,
    )

    xquad_kmeans = run_xquad_variants(
        docs_ranked=docs_ranked,
        subqueries=subqueries_kmeans,
        r_query=r_query,
        r_sub=r_sub_kmeans,
        importances=importances_kmeans,
        tau=100,
        omega=0.5,
    )

    all_runs["xquad_kmeans_uniform"].append(ranking_to_run_df(qid, query, xquad_kmeans["uniform"]))
    all_runs["xquad_kmeans_n"].append(ranking_to_run_df(qid, query, xquad_kmeans["n"]))
    all_runs["xquad_kmeans_redde"].append(ranking_to_run_df(qid, query, xquad_kmeans["redde"]))
    all_runs["xquad_kmeans_crcs"].append(ranking_to_run_df(qid, query, xquad_kmeans["crcs"]))
```

### concaténer les runs

```python
final_runs = {}

for name, run_list in all_runs.items():
    if len(run_list) == 0:
        print(f"Aucune run pour {name}")
        continue

    final_runs[name] = pd.concat(run_list, ignore_index=True)
    print(name, final_runs[name].shape)
```

```python
for name, df_run in final_runs.items():
    df_run.to_csv(f"{name}.csv", index=False)

print("Runs sauvegardées.")
```
